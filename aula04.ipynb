```python
listasEstudantes = []
#----- COMEÇO CADASTRAR ESTUDANTE -----
def cadastrarEstudante(ru):
  print('bem-vindo ao cadastro de estudantes')
  print('O RU do estudante a ser cadastrado é: {}'.format(ru))
  nome = input('Digite o nome do estudante: ')
  turma = input('Digite a turma do estudante: ')
  dicionarioEstudante = {'ru'   : ru,
                        'nome'  : nome,
                        'turma' : turma}

  listasEstudantes.append(dicionarioEstudante.copy())

#----- FIM CADASTRAR ESTUDANTE -----

#----- CONSULTAR ESTUDANTE -----
def consultarEstudante():
  while True:
    try:
      print('Bem-vindo ao consultar estudantes')
      opConsultar = int(input('Entre cm a oção desejada: \n 1 - Consultar todos os estudantes\n 2 - Consultar por Ru\n 3 - Consultar por Turma\n 4 - Retornar'))
      if opConsultar == 1:
        print('Bem-vindo a consultar todos')
        for estudante in listasEstudantes: # Selecionar cada dicionário da minha lista (cada estudante da lista de estudante)
          for key, value in estudante.item.s(): # Selecionar cada conjunto chave/valor do dicionário(ex: 'nome' : Marcela)
            print('{} : {}' .format(key, value))

      elif opConsultar == 2:
        print('Constular RU')
        entrada = int(input('Digite o RU desejado: '))
        for estudante in listasEstudantes:
          if(estudante['ru'] == entrada):
            for key, value in estudante.item.s():
               print('{} : {}' .format(key, value))


      elif opConsultar == 3:
        print('Ben-vindo ao consultar por turma')
        entrada = (input('Digite a turma desejada: ')
        for estudante in listasEstudantes:
          if(estudante['turma'] == entrada):
            for key, value in estudante.item.s():
              print('{} : {}' .format(key, value))

      elif opConsultar == 4:
        return
      else:
        print('digite uma opção existente')
        continue
    except ValueError:
      print('Digite somente números inteiros.')
#----- FIM CONSULTAR ESTUDANTE -----

#----- COMEÇO REMOVER ESTUDANTE -----
def removerEstudante():
  print('bem-vindo ao remover estudante')
  entrada = int((input('Digite o RU desejado: '))
    for estudante in listasEstudantes:
      if(estudante['ru'] == entrada):
       listasEstudantes.remove(estudante)
#----- FIM REMOVER ESTUDANTE ------


#----- COMEÇO MAIN -------

print('Bem-vindo ao programa Controle de Estudantes - Marcela Cristina da Rocha')
registroAcademico = 0
while True:
  try:
    opcao = int(input('Digite a opção desejada:\n 1- Cadastrar Estudante\n 2- Consultar Estudante\n 3- Remover Estudante\n 4- SAIR'))

    if opcao == 1:
      registroAcademico = registroAcademico +1
      cadastrarEstudante(registroAcademico)
    elif opcao == 2:
      consultarEstudante()
    elif opcao == 3:
      removerEstudante()
    elif opcao == 4:
      print('Programa Finalizado')
      break
    else:
      print('pare de digitar números inválidos')
    continue

  except ValueError:
    print('Pare dde digitar valores não inteiros')

#----- FIM MAIN -----

```
